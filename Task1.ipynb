```python
# ============================================================================
# COMPLETE SALES DASHBOARD - NO DOWNLOADS
# ============================================================================

# Step 1: Install required libraries
!pip install pandas numpy matplotlib seaborn plotly openpyxl -q

# Step 2: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("📊 SALES DASHBOARD ANALYSIS")
print("="*60)

# Step 3: Create sample data
print("\n📊 Generating sample data...")

np.random.seed(42)

# Create data
data = {
    'Order_ID': [],
    'Order_Date': [],
    'Category': [],
    'Sub_Category': [],
    'Region': [],
    'Segment': [],
    'Sales': [],
    'Profit': [],
    'Quantity': [],
    'Discount': []
}

categories = ['Furniture', 'Office Supplies', 'Technology']
sub_categories = ['Chairs', 'Tables', 'Binders', 'Paper', 'Phones', 'Accessories']
regions = ['East', 'West', 'North', 'South']
segments = ['Consumer', 'Corporate', 'Home Office']

dates = pd.date_range('2023-01-01', '2025-12-31', freq='D')

for i in range(500):
    data['Order_ID'].append(f'ORD-{i+1:04d}')
    data['Order_Date'].append(np.random.choice(dates))
    data['Category'].append(np.random.choice(categories))
    data['Sub_Category'].append(np.random.choice(sub_categories))
    data['Region'].append(np.random.choice(regions))
    data['Segment'].append(np.random.choice(segments))
    data['Sales'].append(round(np.random.uniform(10, 1000), 2))
    data['Profit'].append(round(np.random.uniform(-50, 200), 2))
    data['Quantity'].append(np.random.randint(1, 10))
    data['Discount'].append(round(np.random.uniform(0, 0.5), 2))

df = pd.DataFrame(data)
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Month_Name'] = df['Order_Date'].dt.strftime('%b')

print(f"✅ Dataset created with {len(df)} records")
print(f"📅 Date Range: {df['Order_Date'].min().date()} to {df['Order_Date'].max().date()}")

# Step 4: Show data preview
print("\n📋 Data Preview (First 5 rows):")
display(df.head())

# Step 5: Data Cleaning
print("\n🧹 Cleaning data...")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicate rows: {df.duplicated().sum()}")

# Step 6: Basic Statistics
print("\n📊 Basic Statistics:")
display(df[['Sales', 'Profit', 'Quantity', 'Discount']].describe())

# Step 7: Pivot Tables
print("\n" + "="*60)
print("📊 PIVOT TABLES")
print("="*60)

# Pivot 1: Sales by Category
pivot1 = pd.pivot_table(df, values='Sales', index='Category', aggfunc='sum')
print("\n1. Total Sales by Category:")
display(pivot1)

# Pivot 2: Sales by Region
pivot2 = pd.pivot_table(df, values='Sales', index='Region', aggfunc='sum')
print("\n2. Total Sales by Region:")
display(pivot2)

# Pivot 3: Profit by Category
pivot3 = pd.pivot_table(df, values='Profit', index='Category', aggfunc='sum')
print("\n3. Total Profit by Category:")
display(pivot3)

# Pivot 4: Sales by Category and Region
pivot4 = pd.pivot_table(df, values='Sales', index='Category', columns='Region', aggfunc='sum', fill_value=0)
print("\n4. Sales by Category and Region:")
display(pivot4)

# Step 8: Conditional Formatting
print("\n" + "="*60)
print("🎯 CONDITIONAL FORMATTING")
print("="*60)

# Category performance
category_performance = df.groupby('Category')['Sales'].sum().reset_index()
category_performance['Performance'] = category_performance['Sales'].apply(
    lambda x: '🟢 HIGH' if x > category_performance['Sales'].median() else '🔴 LOW'
)
print("\nCategory Performance (High/Low):")
display(category_performance)

# Region performance
region_performance = df.groupby('Region')['Profit'].sum().reset_index()
region_performance['Performance'] = region_performance['Profit'].apply(
    lambda x: '🟢 PROFITABLE' if x > 0 else '🔴 LOSS'
)
print("\nRegion Performance (Profit/Loss):")
display(region_performance)

# Step 9: Key Metrics (KPIs)
print("\n" + "="*60)
print("🎯 KEY PERFORMANCE INDICATORS (KPIs)")
print("="*60)

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_profit_margin = (total_profit / total_sales) * 100
total_orders = len(df)
avg_sales = df['Sales'].mean()
avg_discount = df['Discount'].mean()

kpi_data = {
    'Metric': ['Total Sales', 'Total Profit', 'Profit Margin', 'Total Orders', 'Avg Sales/Order', 'Avg Discount'],
    'Value': [
        f'${total_sales:,.2f}',
        f'${total_profit:,.2f}',
        f'{avg_profit_margin:.1f}%',
        f'{total_orders:,}',
        f'${avg_sales:,.2f}',
        f'{avg_discount:.1%}'
    ]
}
kpi_df = pd.DataFrame(kpi_data)
display(kpi_df)

# Step 10: Top Performers
print("\n" + "="*60)
print("🏆 TOP PERFORMERS")
print("="*60)

print("\nTop 3 Categories by Sales:")
display(df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(3))

print("\nTop 3 Sub-Categories by Sales:")
display(df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).head(3))

print("\nTop 3 Regions by Sales:")
display(df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(3))

print("\nTop 3 Segments by Sales:")
display(df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).head(3))

# Step 11: Visualizations
print("\n" + "="*60)
print("📈 VISUALIZATIONS")
print("="*60)

# Chart 1: Sales by Category
plt.figure(figsize=(10, 5))
sales_by_category = df.groupby('Category')['Sales'].sum()
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
bars = plt.bar(sales_by_category.index, sales_by_category.values, color=colors)
plt.title('Total Sales by Category', fontsize=16, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
for bar, val in zip(bars, sales_by_category.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             f'${val:,.0f}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()

# Chart 2: Sales by Region
plt.figure(figsize=(10, 5))
sales_by_region = df.groupby('Region')['Sales'].sum()
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#F9A825']
bars = plt.bar(sales_by_region.index, sales_by_region.values, color=colors)
plt.title('Total Sales by Region', fontsize=16, fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
for bar, val in zip(bars, sales_by_region.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             f'${val:,.0f}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()

# Chart 3: Profit by Category
plt.figure(figsize=(10, 5))
profit_by_category = df.groupby('Category')['Profit'].sum()
colors_profit = ['#2ECC71' if x > 0 else '#E74C3C' for x in profit_by_category.values]
bars = plt.bar(profit_by_category.index, profit_by_category.values, color=colors_profit)
plt.title('Total Profit by Category', fontsize=16, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Total Profit ($)')
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
for bar, val in zip(bars, profit_by_category.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
             f'${val:,.0f}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()

# Chart 4: Monthly Sales Trend
plt.figure(figsize=(12, 5))
monthly_sales = df.groupby('Month_Name')['Sales'].sum()
monthly_sales = monthly_sales.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=2,
         markersize=8, color='#2E86AB')
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True, alpha=0.3)
for i, val in enumerate(monthly_sales.values):
    plt.text(i, val + 5, f'${val:,.0f}', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

# Chart 5: Sales vs Profit Scatter
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Profit'], alpha=0.5, c=df['Discount'], cmap='RdYlGn', s=50)
plt.colorbar(label='Discount Rate')
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)
plt.axvline(x=df['Sales'].mean(), color='blue', linestyle='--', alpha=0.5)
plt.title('Sales vs Profit Analysis', fontsize=16, fontweight='bold')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Chart 6: Pie Chart - Sales by Segment
plt.figure(figsize=(8, 8))
segment_sales = df.groupby('Segment')['Sales'].sum()
colors = ['#FF9F1C', '#2EC4B6', '#E71D36']
plt.pie(segment_sales.values, labels=segment_sales.index, autopct='%1.1f%%',
        colors=colors, startangle=90, explode=(0.05, 0.05, 0.05))
plt.title('Sales Distribution by Segment', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Step 12: Interactive Dashboard
print("\n" + "="*60)
print("📊 INTERACTIVE DASHBOARD (Plotly)")
print("="*60)

# Create subplots
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Sales by Category', 'Sales by Region',
                    'Profit by Category', 'Monthly Sales Trend'),
    specs=[[{'type': 'bar'}, {'type': 'bar'}],
           [{'type': 'bar'}, {'type': 'scatter'}]]
)

# 1. Sales by Category
cat_sales = df.groupby('Category')['Sales'].sum()
fig.add_trace(
    go.Bar(x=cat_sales.index, y=cat_sales.values, name='Sales by Category',
           marker_color=['#FF6B6B', '#4ECDC4', '#45B7D1']),
    row=1, col=1
)

# 2. Sales by Region
reg_sales = df.groupby('Region')['Sales'].sum()
fig.add_trace(
    go.Bar(x=reg_sales.index, y=reg_sales.values, name='Sales by Region',
           marker_color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#F9A825']),
    row=1, col=2
)

# 3. Profit by Category
cat_profit = df.groupby('Category')['Profit'].sum()
fig.add_trace(
    go.Bar(x=cat_profit.index, y=cat_profit.values, name='Profit by Category',
           marker_color=['#2ECC71' if x > 0 else '#E74C3C' for x in cat_profit.values]),
    row=2, col=1
)

# 4. Monthly Sales Trend
monthly = df.groupby('Month_Name')['Sales'].sum()
monthly = monthly.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
fig.add_trace(
    go.Scatter(x=monthly.index, y=monthly.values, mode='lines+markers',
               name='Monthly Trend', line=dict(color='#2E86AB', width=3),
               marker=dict(size=8, color='#E84855')),
    row=2, col=2
)

# Update layout
fig.update_layout(
    height=700,
    width=900,
    title_text="📊 Sales Dashboard",
    title_font_size=20,
    showlegend=False,
    template='plotly_white'
)

fig.show()

# Step 13: Final Summary
print("\n" + "="*60)
print("📋 EXECUTIVE SUMMARY")
print("="*60)

print(f"""
📊 Dashboard Summary:
• Total Sales: ${total_sales:,.2f}
• Total Profit: ${total_profit:,.2f}
• Profit Margin: {avg_profit_margin:.1f}%
• Total Orders: {total_orders:,}
• Average Order Value: ${avg_sales:,.2f}

🏆 Top Category: {df.groupby('Category')['Sales'].sum().idxmax()}
   (${df.groupby('Category')['Sales'].sum().max():,.2f})

📍 Best Region: {df.groupby('Region')['Sales'].sum().idxmax()}
   (${df.groupby('Region')['Sales'].sum().max():,.2f})

📈 Best Month for Sales: {monthly.idxmax()}
   (${monthly.max():,.2f})
""")

print("✅ Dashboard Complete! 🎉")
print("="*60)
```
